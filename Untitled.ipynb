## Visualization
Now we use different visualization tools to check different aspects of Supermarket sales.

```python
#Gender Count
plt.figure(figsize=(10,7))
plt.style.use('fivethirtyeight')
ax= sns.countplot(x = 'Gender', data=df , palette = "copper" )
ax.set_xlabel(xlabel= "Gender",fontsize=18)
ax.set_ylabel(ylabel = "Gender count", fontsize = 18)
ax.set_title(label = "Gender count in supermarket", fontsize = 20)
plt.show()
```

Here we can see that the number of males and females entering the store is almost equal. But the visualization looks suspicious. Let’s check numeric data.

```python
df.groupby(['Gender']). agg({'Total':'sum'})
```

```python
#Customer type

plt.style.use('ggplot')
plt.figure(figsize= (14,6))
ax = sns.countplot(x = "Customer type", data = df, palette = "rocket_r")
ax.set_title("Type of customers", fontsize = 25)
ax.set_xlabel("Customer type", fontsize = 16)
ax.set_ylabel("Customer Count", fontsize = 16)
plt.show()
```

```python
df.groupby(['Customer type']). agg({'Total':'sum'})
```

Above we can see the type of customer in all branch combined now let’s check for different branch.

```python
plt.figure(figsize=(14,6))
plt.style.use('classic')
ax = sns.countplot(x = "Customer type", hue = "Branch", data = df, palette= "rocket_r")
ax.set_title(label = "Customer type in different branch", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Customer Count", fontsize = 16)
```

```python
#Checking the different payment methods used.

plt.figure(figsize = (14,6))
ax = sns.countplot(x = "Payment", data = df, palette = "tab20")
ax.set_title(label = "Payment methods of customers ", fontsize= 25)
ax.set_xlabel(xlabel = "Payment method", fontsize = 16)
ax.set_ylabel(ylabel = " Customer Count", fontsize = 16)
plt.show()
```

```python
#Payment method distribution in all branches
plt.figure(figsize = (14,6))
plt.style.use('classic')
ax = sns.countplot(x="Payment", hue = "Branch", data = df, palette= "tab20")
ax.set_title(label = "Payment distribution in all branches", fontsize= 25)
ax.set_xlabel(xlabel = "Payment method", fontsize = 16)
ax.set_ylabel(ylabel = "Peple Count", fontsize = 16)
plt.show()
```

```python
#the rating distribution in 3 branches
plt.figure(figsize=(14,6)) 
ax = sns.boxplot(x="Branch", y = "Rating" ,data =df, palette= "RdYlBu")
ax.set_title("Rating distribution between branches", fontsize = 25)
ax.set_xlabel(xlabel = "Branches", fontsize = 16)
ax.set_ylabel(ylabel = "Rating distribution", fontsize = 16)
plt.show()
```

We can see that the average rating of branch A and C is more than seven and branch B is less than 7.

```python
#Max sales time
df["Time"]= pd.to_datetime(df["Time"])
df["Hour"]= (df["Time"]).dt.hour
plt.figure(figsize=(14,6))
plt.style.use('classic')
SalesTime = sns.lineplot(x="Hour", y ="Quantity", data = df).set_title("product sales per Hour")
plt.show()
```

```python
#Rating vs Sales
plt.figure(figsize=(14,6))
plt.style.use('classic')
rating_vs_sales = sns.lineplot(x="Total", y= "Rating", data=df)
```

```python
sns.scatterplot(x="Total", y="Rating", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.style.use('classic')
ax = sns.boxenplot(x = "Quantity", y = "Product line", data = df,palette = "copper")
ax.set_title(label = "Average sales of different lines of products", fontsize = 25)
ax.set_xlabel(xlabel = "Qunatity Sales",fontsize = 16)
ax.set_ylabel(ylabel = "Product Line", fontsize = 16)
plt.show()
```

Here we can see that the average sales of different lines of products. Health and beauty making the highest sales whereas Fashon accessories making the lowest sales.

```python
#the sales count of these products.

plt.figure(figsize=(14,6))
ax = sns.countplot(y='Product line', data=df, order = df['Product line'].value_counts().index , palette = "RdBu")
ax.set_title(label = "Sales count of products", fontsize = 25)
ax.set_xlabel(xlabel = "Sales count", fontsize = 16)
ax.set_ylabel(ylabel= "Product Line", fontsize = 16)
plt.show()
```

```python
#Total sales of product using boxenplot
plt.figure(figsize=(14,6))
plt.style.use('classic')
ax = sns.boxenplot(y= "Product line", x= "Total", data = df,palette = "Spectral")
ax.set_title(label = " Total sales of product", fontsize = 25)
ax.set_xlabel(xlabel = "Total sales", fontsize = 16)
ax.set_ylabel(ylabel = "Product Line", fontsize = 16)
```

```python
# average ratings of products

plt.figure(figsize = (14,6))
plt.style.use('classic')
ax = sns.boxenplot(y = "Product line", x = "Rating", data = df , palette = "coolwarm")
ax.set_title("Average rating of product line", fontsize = 25)
ax.set_xlabel("Rating", fontsize = 16)
ax.set_ylabel("Product line", fontsize = 16)
plt.show()
```

```python
#Product sales on the basis of gender
plt.style.use('classic')
plt.figure(figsize = (14,6))
ax= sns.stripplot(y= "Product line", x = "Total", hue = "Gender", data = df , jitter=True)
ax.set_title(label = "Product sales on the basis of gender")
ax.set_xlabel(xlabel = " Total sales of products")
ax.set_ylabel(ylabel = "Product Line")
plt.show()
```

```python
#Product and gross income
plt.figure(figsize = (14,6))
plt.style.use('classic')
ax = sns.relplot(y= "Product line", x = "gross income", data = df)
plt.show()
```

# Project Insights: Supermarket Sales Analysis
## Dataset Overview
### The dataset contains 1000 rows and 17 columns.
### There are no missing values, so the dataset is clean and suitable for analysis.
### Data includes customer details, product categories, sales, and ratings.
## Gender Analysis
### The number of male and female customers is almost equal.
### However, female customers contribute slightly more to total sales than males.
## Customer Type Insights
### The number of Member and Normal customers is nearly equal.
### Members generate slightly higher total revenue, indicating better engagement or spending behavior.
### Across branches, customer distribution remains fairly balanced.
## Branch Performance
### All three branches show similar customer distribution.
### Branch A and C have higher average ratings (>7) compared to Branch B (<7).
### This suggests better customer satisfaction in Branch A and C.
## Payment Methods
### Customers use multiple payment methods, with no extreme dominance of a single method.
### Payment preferences are consistent across all branches.
## Sales Timing (Important Insight ⭐)
### Maximum sales occur around 2 PM (14:00 hrs).
### This indicates peak shopping hours during mid-day.
## Rating vs Sales
### There is no strong relationship between total sales and ratings.
### High sales do not necessarily mean high customer satisfaction.
## Product Line Analysis
### Health and Beauty products generate the highest sales.
### Fashion Accessories have the lowest sales.
### Some product categories are clearly more popular than others.
## Sales Count of Products
### Certain product lines have higher purchase frequency, indicating strong demand.
### Popular products drive majority of transactions.
## Total Sales by Product Line
### Some product categories contribute significantly more to total revenue.
### Sales distribution varies across product lines.
## Product Ratings
### Most product lines have ratings above average (~7).
### Indicates generally good customer satisfaction across products.
## Gender-Based Product Sales
### Both genders purchase across all product lines.
### Some variation exists, but no strong gender dominance for specific products.
## Gross Income Insights
### Gross income varies across product lines.
### Certain products contribute more profit despite similar sales volume.
## Final Conclusion
1. The supermarket has balanced customer demographics.
2. Peak sales occur in the afternoon, useful for business planning.
3. Customer satisfaction is generally good, but not directly linked to sales.
4.  Product line performance varies significantly, with some categories driving most revenue.
5. Membership programs slightly improve revenue generation.
